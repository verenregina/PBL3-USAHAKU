**IMPORT LIBRARY + LOAD DATA**

```python
# Import library
import pandas as pd
import numpy as np

# Upload file
from google.colab import files
uploaded = files.upload()

# Load dataset (sesuaikan nama file)
df = pd.read_excel('dataset_umkm_01.xlsx')

# Tampilkan data
df.head()
```

**PREPROCESSING DATA**

```python
# contoh kategorisasi
def kategori_aset(x):
    if x < 10000000:
        return 'kecil'
    elif x < 50000000:
        return 'sedang'
    else:
        return 'besar'

def kategori_omset(x):
    if x < 5000000:
        return 'rendah'
    elif x < 20000000:
        return 'sedang'
    else:
        return 'tinggi'

#terapkan
df['aset_kat'] = df['aset'].apply(kategori_aset)
df['omset_kat'] = df['omset'].apply(kategori_omset)

# label (target)
def kelayakan(row):
    if row['aset_kat'] == 'besar' and row['omset_kat'] == 'tinggi':
        return 'layak'
    else:
        return 'tidak_layak'

df['kelayakan'] = df.apply(kelayakan, axis=1)

df.head()
```

```python
from IPython.display import display, Markdown

display(Markdown("""
## Penjelasan Hasil Kategorisasi dan Perhitungan

Pada tabel di atas, terdapat penambahan tiga kolom baru yaitu **aset_kat**, **omset_kat**, dan **kelayakan**.
Ketiga kolom tersebut merupakan hasil dari proses transformasi data numerik menjadi data kategorikal
berdasarkan aturan (rule) yang telah ditentukan sebelumnya.

### 1. Kategorisasi Aset (aset_kat)
Nilai aset yang semula berupa angka diubah menjadi kategori sebagai berikut:
- Aset < 10.000.000 → **kecil**
- Aset 10.000.000 – 50.000.000 → **sedang**
- Aset > 50.000.000 → **besar**

Contoh:
- Data ke-0 memiliki aset sebesar 5.497.149 → termasuk kategori **kecil**

---

### 2. Kategorisasi Omset (omset_kat)
Nilai omset juga dikategorikan menjadi:
- Omset < 5.000.000 → **rendah**
- Omset 5.000.000 – 20.000.000 → **sedang**
- Omset > 20.000.000 → **tinggi**

Contoh:
- Data ke-0 memiliki omset sebesar 3.347.794 → termasuk kategori **rendah**

---

### 3. Penentuan Kelayakan (kelayakan)
Kelayakan usaha ditentukan berdasarkan aturan sederhana:
- Jika **aset = besar** DAN **omset = tinggi** → **layak**
- Selain itu → **tidak_layak**

Contoh:
- Data ke-0 memiliki aset = kecil dan omset = rendah
  → tidak memenuhi syarat
  → hasil: **tidak_layak**

---

### Kesimpulan
Perhitungan ini bukan berupa operasi matematis kompleks, melainkan proses **klasifikasi berbasis aturan (rule-based classification)**.
Data numerik diubah menjadi kategori agar lebih mudah digunakan dalam metode seperti Forward Chaining, Backward Chaining,
Naive Bayes, dan Certainty Factor.

Dengan adanya kategorisasi ini, sistem dapat mengambil keputusan secara lebih terstruktur dan sesuai dengan logika yang telah ditentukan.
"""))
```

**BAYES**

```python
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.preprocessing import LabelEncoder

# Encode data
le = LabelEncoder()
df['aset_enc'] = le.fit_transform(df['aset_kat'])
df['omset_enc'] = le.fit_transform(df['omset_kat'])
df['target'] = le.fit_transform(df['kelayakan'])

# Feature & target
X = df[['aset_enc', 'omset_enc']]
y = df['target']

# Split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

# Model
model = GaussianNB()
model.fit(X_train, y_train)

# Prediksi
y_pred = model.predict(X_test)

# Akurasi
from sklearn.metrics import accuracy_score
print("Akurasi:", accuracy_score(y_test, y_pred))
```

Dalam penelitian ini, Teorema Bayes digunakan untuk menentukan kelayakan usaha UMKM berdasarkan beberapa variabel, seperti aset dan omset. Data yang digunakan terlebih dahulu dikategorikan, kemudian dihitung probabilitas masing-masing kelas (layak dan tidak layak).

Proses perhitungan dilakukan dengan cara mengalikan probabilitas setiap atribut terhadap kelas tertentu, kemudian dibandingkan dengan kelas lainnya. Kelas dengan nilai probabilitas terbesar akan dipilih sebagai hasil prediksi.

Sebagai contoh, jika suatu UMKM memiliki aset besar dan omset tinggi, maka akan dihitung probabilitas usaha tersebut termasuk dalam kategori "layak" dan "tidak layak". Hasil dengan nilai probabilitas terbesar akan menjadi keputusan akhir.

Metode ini memiliki beberapa kelebihan, antara lain mampu mengolah data dengan cepat, sederhana, dan efektif untuk klasifikasi. Namun, metode ini juga memiliki kelemahan yaitu mengasumsikan bahwa setiap atribut bersifat independen, yang dalam kenyataannya tidak selalu demikian.

Dengan demikian, Teorema Bayes dapat digunakan sebagai salah satu pendekatan dalam membantu pengambilan keputusan terkait kelayakan usaha UMKM secara objektif dan berbasis data.
